```python
## Dataset with  690 obs. of  15 features:
##  $ f0 : Factor w/ 2 levels "a","b" (Male)
##  $ f1 : num  30.8 58.7 24.5 27.8 20.2 ..(Age)
##  $ f2 : num  0 4.46 0.5 1.54 5.62.. (Debt)
##  $ f3 : Factor w/ 3 levels "l","u","y" (Married)
##  $ f4 : Factor w/ 3 levels "g","gg","p" (BankCustomer)
##  $ f5 : Factor w/ 14 levels "aa","c","cc" (EducationLevel)
##  $ f6 : Factor w/ 9 levels "bb","dd","ff" (Ethnicity)
##  $ f7 : num  1.25 3.04 1.5 3.75 1.71... (YearsEmployed)
##  $ f8 : Factor w/ 2 levels "f","t" (PriorDefault)
##  $ f9: Factor w/ 2 levels "f","t" (Employed)
##  $ f10: int  1 6 0 5 0 0 0 0 0 0 ... (CreditScore)
##  $ f11: Factor w/ 2 levels "f","t" (DriversLicense)
##  $ f12: Factor w/ 3 levels "g","p","s" (Citizen   )
##  $ f13: int  202 43 280 100 120 360 164 80 180 52 ...(Zipcode)
##  $ f14: int  0 560 824 3 0 0 31285 1349 314 1442 ... (Income)
##  $ f15: Factor w/ 2 levels "-","+"  (Approved)
```

```python
import numpy as np
import pandas as pd
```

```python
rawdata = pd.read_csv('Dataset/crx.data',header=None)
```

```python
rawdata.head()
```

```python
# Removing the missing values
miss_index = []
for col in rawdata.columns:
    miss_index.extend(list(rawdata[rawdata[col]=="?"].index.values))
    
rawdata_2 = rawdata.drop(miss_index, axis=0)
```

```python
# Number of Missing Values
n_missval = len(rawdata)-len(rawdata_2)
```

```python
from collections import defaultdict
from sklearn.preprocessing import LabelEncoder, scale, MinMaxScaler

d = defaultdict(LabelEncoder)

# Encoding the necessary features(0,3,4,5,6,8,9,11,12,15)
enc = rawdata_2[[0,3,4,5,6,8,9,11,12,15]].apply(lambda x: d[x.name].fit_transform(x))
```

```python
rawdata_2[15].value_counts()
```

```python
# Converting to float numbers

rawdata_2 = rawdata_2[[1,2,7,10,13,14]].astype(float)
```

```python
rawdata_2.head()
```

```python
dataset_1 = pd.concat([rawdata_2,enc],axis=1,sort=True)
dataset_1.head()
```

```python
datase_1 = dataset_1.reindex(sorted(dataset_1.columns), axis=1).values
```

```python
# Normalization of the data
min_max_scaler = MinMaxScaler()
dataset = min_max_scaler.fit_transform(dataset_1)
```

```python
pd.DataFrame(dataset)
```

```python
from sklearn.model_selection import KFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
kf = KFold(n_splits=10, shuffle=True, random_state=10)
```

```python
dtree = DecisionTreeClassifier()
```

```python
acc = []
precision = []
recall = []
F1 = []

for train_index, test_index in kf.split(dataset):
    train, test = dataset[train_index], dataset[test_index]
    
    x_train, x_test = train[:,:14], test[:,:14]
    y_train, y_test = train[:,15], test[:,15]
    
    dtree.fit(x_train,y_train)
    predictions = dtree.predict(x_test)
    TN, FP, FN, TP = confusion_matrix(y_test,predictions).ravel()
    
    acc_fold = (TP+TN)/(TP+FP+TN+FN)
    precision_fold = TP/(TP+FP)
    recall_fold = TP/(TP+FN)
    F1_fold = 2*(precision_fold * recall_fold)/(precision_fold + recall_fold)
    
    acc.append(acc_fold)
    precision.append(precision_fold)
    recall.append(recall_fold)
    F1.append(F1_fold)
    
#     print(con_mat)
#     print(TP)
    print('\n')
    print(classification_report(y_test,predictions))
```

```python
acc_avg =  np.average(acc)
precision_avg =  np.average(precision)
recall_avg =  np.average(recall)
F1_avg = np.average(F1)
```
