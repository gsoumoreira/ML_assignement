```python
## Dataset with  690 obs. of  15 features:
##  $ f0 : Factor w/ 2 levels "a","b" (Male)
##  $ f1 : num  30.8 58.7 24.5 27.8 20.2 ..(Age)
##  $ f2 : num  0 4.46 0.5 1.54 5.62.. (Debt)
##  $ f3 : Factor w/ 3 levels "l","u","y" (Married)
##  $ f4 : Factor w/ 3 levels "g","gg","p" (BankCustomer)
##  $ f5 : Factor w/ 14 levels "aa","c","cc" (EducationLevel)
##  $ f6 : Factor w/ 9 levels "bb","dd","ff" (Ethnicity)
##  $ f7 : num  1.25 3.04 1.5 3.75 1.71... (YearsEmployed)
##  $ f8 : Factor w/ 2 levels "f","t" (PriorDefault)
##  $ f9: Factor w/ 2 levels "f","t" (Employed)
##  $ f10: int  1 6 0 5 0 0 0 0 0 0 ... (CreditScore)
##  $ f11: Factor w/ 2 levels "f","t" (DriversLicense)
##  $ f12: Factor w/ 3 levels "g","p","s" (Citizen   )
##  $ f13: int  202 43 280 100 120 360 164 80 180 52 ...(Zipcode)
##  $ f14: int  0 560 824 3 0 0 31285 1349 314 1442 ... (Income)
##  $ f15: Factor w/ 2 levels "-","+"  (Approved)
```

```python
import numpy as np
import pandas as pd
```

```python
rawdata = pd.read_csv('Dataset/crx.data',header=None)
```

```python
37/690*100
```

```python
rawdata.head()
```

```python
# Removing the missing values
miss_index = []
for col in rawdata.columns:
    miss_index.extend(list(rawdata[rawdata[col]=="?"].index.values))
    
data = rawdata.drop(miss_index, axis=0)
```

```python
# Number of Missing Values
n_missval = len(rawdata)-len(data)
```

```python
from collections import defaultdict
from sklearn.preprocessing import LabelEncoder

d = defaultdict(LabelEncoder)

# Encoding the necessary features(0,3,4,5,6,8,9,11,12,15)
enc = data[[0,3,4,5,6,8,9,11,12,15]].apply(lambda x: d[x.name].fit_transform(x))
```

```python
enc
```

```python
data[15].value_counts()
```

```python
# Converting to float numbers

data = data[[1,2,7,10,13,14]].astype(float)
```

```python
data.head()
```

```python
data3 = pd.concat([data,enc],axis=1,sort=True)
data3.head()
```

```python
data3 = data3.reindex(sorted(data3.columns), axis=1).values
```

```python
data3
```

```python
data3
```

```python
from sklearn.model_selection import KFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
kf = KFold(n_splits=10, shuffle=True, random_state=10)
```

```python
dtree = DecisionTreeClassifier()
```

```python
for train_index, test_index in kf.split(data3):
    train, test = data3[train_index], data3[test_index]
    
    x_train, x_test = train[:,:14], test[:,:14]
    y_train, y_test = train[:,15], test[:,15]
    
    dtree.fit(x_train,y_train)
    predictions = dtree.predict(x_test)
    print(confusion_matrix(y_test,predictions))
    print('\n')
    print(classification_report(y_test,predictions))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=200)
```

```python
for train_index, test_index in kf.split(data3):
    train, test = data3[train_index], data3[test_index]
    
    x_train, x_test = train[:,:14], test[:,:14]
    y_train, y_test = train[:,15], test[:,15]
    
    rfc.fit(x_train,y_train)
    predictions = rfc.predict(x_test)
    print(confusion_matrix(y_test,predictions))
    print('\n')
    print(classification_report(y_test,predictions))
```
